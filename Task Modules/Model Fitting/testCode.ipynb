```python
# IPython is what you are using now to run the notebook
import IPython
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../Data/listings.csv')
```

```python
data.head()
```

```python
sns.set_style('darkgrid')
sns.countplot(data['room_type'])

## In the plot, we can see that the "Entire home/apt" is the most favorite room type which is above 25000 
## while "Private room" is also quite popular, which is around 22500. Comparing to these two room types,
## only a small amount of people choose "Shared room" and "Hotel room".
```

```python
sns.scatterplot(x = data["number_of_reviews"], y = data["price"])
plt.title('Price vs. Number of reviews')
## In the scatterplot, we can see that as the number of reviews increases, the price actually decreases. 
## One possible reason for this is that when the price of the room is really high, a few people choose this
## type of room. 
```

```python
sns.scatterplot(x = data["reviews_per_month"], y = data["price"])
plt.title('Price vs. Reviews per month')
## In the scatterplot, we can see that as the number of reviews per month increases, the price actually decreases. 
## This pattern is quite similar to last scatterplot. This makes sense since the same reason can apply to this
## plot. 
```
